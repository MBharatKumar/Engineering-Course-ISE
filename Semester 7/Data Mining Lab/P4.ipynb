```python
import pandas as pd
import numpy as np
import pyfpgrowth as fpg
```

```python
df = pd.read_csv('transactions.csv')
```

```python
df.head()
```

```python
df1 = df.drop(['TID'], axis=1)
```

```python
df1.head()
```

```python
df1 = df1.where(df1 != 1, df1.columns.to_series(), axis=1)
df1 = df1.where(df1 != 0, np.nan, axis=1)
```

```python
df1.head()
```

```python
transactions = df1.stack().groupby(level=0).apply(list).tolist()
```

```python
transactions[0]
```

Frequent Itemsets

```python
freq_itemsets = fpg.find_frequent_patterns(transactions, 0.03)
```

```python
freq_itemsets
```

Association Rules

```python
fpg.generate_association_rules(freq_itemsets, 0.7)
```
